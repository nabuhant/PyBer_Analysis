```python
# Import dependencies.
import matplotlib.pyplot as plt 
#chaining matplotlib.pyplot is equivalent to from matplotlib import pyplot
```

```python
%matplotlib inline
# inline value displays the chart within the cell for a Jupyter Notebook file
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

###### Create a Line Chart Using the MATLAB Method

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

###### Create a Line Chart Using the Object-Oriented Interface

```python
# Create the plot with ax.plt() (method 1)
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() (method 2) where 
# fig, ax = plt.subplots() becomes:
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# To create a quick graph using the object-oriented interface method, use the following two lines:
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

###### Use plt.show()

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

###### Annotate Charts Using the MATLAB Method

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis,marker = "*", color = "blue", linewidth = 2, label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()

# Add a grid
plt.grid()
```

###### Annotate Charts Using the Object-Oriented Interface

```python
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
fig, ax = plt.subplots()
ax.plot (x_axis, y_axis, color = "green", marker = "D", linewidth = 2, label = "Boston")

# Add a legend for the city of Boston.
ax.legend()

# Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')

# Add a y-axis limit.
ax.set_ylim(0, 45)

# Add grid lines.
ax.grid()
```

### 5.1.5 Create Bar Charts Using the MATLAB Approach

```python
# Create a bar plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.gca().invert_yaxis()
plt.legend()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
```

### 5.1.6 Create Bar Charts Using the Object-Oriented Approach

```python
# Create the vertical bar plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label='New York')
```

```python
# Create the horizontal bar plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
ax.legend()
# Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.invert_yaxis()
```

### 5.1.6 Create Scatter Plots and Bubble Charts

###### Create a Scatter Plot Using the MATLAB Method

```python
# using the plt.plot() method
plt.plot(x_axis, y_axis, 'o')
```

```python
# using the plt.scatter() method
plt.scatter(y_axis, x_axis, c="red", label = "Chicago")
plt.gca().invert_yaxis()
plt.legend()
plt.xlim(0, 50)
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
```

###### Create a Bubble Chart Using the MATLAB Method

```python
plt.scatter(x_axis, y_axis, s=y_axis)
# the s (size) parameter is equal to a feature of our data (y-axis data)
```

```python
# make the markers larger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    #y_axis_larger = data*3 ## WRONG WRONG Wrong
print(y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Use list comprehension instead of for loop above
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

###### Create a Scatter Plot Using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

###### Create a Bubble Chart Using the Object-Oriented Interface

```python
fig, ax = plt.subplots()
ax.scatter(y_axis,
           x_axis,
           c = "skyblue",
           s=[z * 5 for z in y_axis],
           edgecolors="black",
           linewidth = 2,
           alpha = 0.2,
           label = "Boston")

ax.set_title("PyBer Fare by Month")
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.invert_yaxis()
ax.set_xlim(0,50)
ax.legend()
```

### 5.1.8 Create Pie Charts

###### Create a Pie Chart Using the MATLAB Method

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0) # a tuple qual in elemnts as the no. of wedges

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.subplots(figsize=(8, 8)) #resize needs to be first line
plt.pie(y_axis, 
        explode=explode_values,
        labels=x_axis,
        colors=colors,
        autopct='%.1f%%')

#The % before and after the .1f% formats the number as a percentage
#autopct parameter adds percentage values
plt.show()
```

###### Create a Pie Chart Using the Object-Oriented Interface

```python
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0) # a tuple qual in elemnts as the no. of wedges

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, 
        explode=explode_values,
        labels=x_axis,
        colors=colors,
        autopct='%.1f%%',
        shadow = True,
        counterclock = False,
        startangle = 95
        )
plt.show()
```
